```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python

def monod_simulation2(t, y, vmax, km, y_coeff, m):
    x, s = y
    mu = vmax * s / (km + s)
    dxdt = mu * x - (m * x)
    dsdt = -1/y_coeff * (mu * x)
    return (dxdt, dsdt)

  # specific growth rate
X0 = 0.1
S0 = 50
km = 0.5
y_coeff = 0.4
vmax = 0.8
m = 0.1
t_span = (0, 10)  # time span for the integration
 

solution = solve_ivp(monod_simulation2, t_span, [X0, S0], args=(vmax, km, y_coeff, m))

    
t = solution.t
X = solution.y[0]
S = solution.y[1]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, X, label='Biomass concentration (X)', color='b')
plt.plot(t, S, label='Substrate concentration (S)', color='g')
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (g/L)')
plt.legend()
plt.title('Biomass and Substrate Concentrations Over Time')
plt.grid(True)
plt.show()
```
